```python


## 문자열 압축

def solution(s):

    N = len(s)
    ans = float("inf")

    for r in range(1,N + 1,1):
        index = 0
        new_str = ""
        while(1):   # 한 문자 압축하기
            if index >= N:  break
            elif index + 2*r - 1 >= N:
                new_str += s[index:]
                break
            n_rep = 1
            chunk = s[index: index + r]
            while(1):       # 한 문구 반복
                if index + r - 1 >= N: break  
                elif chunk == s[index + r : index + 2*r]:
                    n_rep += 1
                    index += r
                else:
                    index += r
                    break

            if n_rep == 1:
                new_str += chunk
            else: 
                new_str = new_str + str(n_rep) + chunk
        if len(new_str) < ans:
            ans = len(new_str)

    return ans
```

```python

## 괄호변환

def check(str):
    correct = True
    balanced = False
    left = 0;   right = 0
    for s in str:
        if s == "(": 
            left += 1
        else:
            right += 1
        if right > left:
            correct = False
    if left == right:
        balanced = True
    else:
        balanced = False
    return correct , balanced

def divide(str):
    u = "";     v = ""
    for index in range(len(str)):
        u += str[index]
        c , b = check(u)
        if b == True:
            v += str[index + 1 :]
            break
    print("분리!" , u , v)
    return u , v

def solution(p):
    if p == "":
        return ""
    answer = ''
    u,v = divide(p)
    c , b = check(u)
    v = solution(v)
    if c == 1:
        answer = u + v
    else:
        new_u = ""
        for j in range(len(u)):
            if j == 0 or j == len(u) - 1:   continue
            if u[j] == "(":
                new_u += ")"
            else:
                new_u +="("
        answer = "("+  v + ")" + new_u
    return answer

```

```python

## 자물쇠와 열쇠

import copy

def rotate(arr):
    arr_copy = copy.deepcopy(arr)
    #arr_copy = [[-1]*len(arr)]*len(arr)
    for x in range(len(arr)):
        for y in range(len(arr)):
            r_x , r_y = y , len(arr) - x - 1
            arr_copy[r_x][r_y] = arr[x][y] 
    return arr_copy
    
def check_right(key , lock , ref , M , N):
    
    for x in range(N):
        for y in range(N):         
            # 키값 확인
            k_x = x + ref[0]; k_y = y + ref[1]      
            if k_x < 0 or k_x >= M or k_y <0 or k_y >= M:   key_stat = 0
            else: key_stat = key[x + ref[0]][y + ref[1]]
            # 맞물리는지 확인
            if lock[x][y] ^ key_stat == 0:
                return False
    return True

def solution(key, lock): 
    M = len(key[0]);    N = len(lock[0])
            
    next_key = copy.deepcopy(key)
    for _ in range(4):              # 회전 
        next_key = rotate(next_key)
        for x in range(M-1 , 1-N-1,-1):     # 좌표 이동
            for y in range(M-1 ,1-N-1 , -1):
                ref = [x,y]
                if check_right(next_key,lock,ref,M,N) == True:
                    return True
    return False
```

```python

## 기둥과 보 설치

ans = []
import  copy

def check_possible(build_pos , build_type , state) :
    N = len(state[0][0])
    if build_type == 0: # 벽
        if build_pos[1] == 0:
            return True
        if 0 <= build_pos[1] - 1< N and state[build_type][build_pos[0]][build_pos[1] - 1] == 1:
            return True
        if 0 <= build_pos[0] - 1< N and state[1][build_pos[0] - 1][build_pos[1]] == 1:
            return True
        if state[1][build_pos[0]][build_pos[1]] == 1:
            return True
    else:   # 보
        if 0 <= build_pos[1] - 1 < N and state[0][build_pos[0]][build_pos[1] - 1] == 1 :
            return True
        if 0 <= build_pos[0] + 1 < N and 0 <= build_pos[1] - 1 < N and state[0][build_pos[0] + 1][build_pos[1] - 1] == 1 :
            return True
        if 0 <= build_pos[1] - 1 < N and 0 <= build_pos[1] + 1 < N:
            if state[1][build_pos[0] - 1][build_pos[1]] == 1 and state[1][build_pos[0] + 1][build_pos[1]] == 1 :
                return True
    return False
 
def build(order , state):
    build_pos = order[0:2]
    build_type = order[2]
    build_opt = order[3]
    if build_opt == 1:   # 건설
        if state[build_type][build_pos[0]][build_pos[1]] == 1:
            return
        if check_possible(build_pos , build_type , state) == True:
            state[build_type][build_pos[0]][build_pos[1]] = 1
            ans.append([build_pos[0] , build_pos[1] , build_type])
            print("건설 : {} , x : {} , y : {}".format(build_type , build_pos[0] , build_pos[1]))
    else:           # 삭제
        remove_flag = 1
        new_state = copy.deepcopy(state)
        new_state[build_type][build_pos[0]][build_pos[1]] = 0        
        for i in range(len(ans)):
            n_pos = ans[i][0:2];  n_type = ans[i][2]
            if build_pos == n_pos and build_type == n_type:
                continue
            if check_possible( n_pos , n_type , new_state) == False:
                remove_flag = 0;    break
        if remove_flag == 1:
            state[build_type][build_pos[0]][build_pos[1]] = 0
            ans.pop(ans.index([build_pos[0] , build_pos[1] , build_type]))
            print("삭제완료")

def solution(n, build_frame):
    state = [[ [0 for _ in range(n + 1)] for _ in range(n + 1) ] for _ in range(2)]
    for i in range(len(build_frame)):
        build(build_frame[i] , state)
    return sorted(ans, key = lambda x : (x[0], x[1] , x[2]))
```
